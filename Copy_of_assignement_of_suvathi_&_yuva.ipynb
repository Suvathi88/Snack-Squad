```python
import webrtcvad
import wave
import contextlib

def get_speech_regions(file_path):
    vad = webrtcvad.Vad(1)  # Aggressiveness: 0-3
    with contextlib.closing(wave.open(file_path, 'rb')) as wf:
        sample_rate = wf.getframerate()
        channels = wf.getnchannels()
        if channels != 1 or sample_rate != 16000:
            raise ValueError("Only 16kHz mono WAV supported for VAD.")
        frames = wf.readframes(wf.getnframes())
    return frames  # Placeholder for region detection
```

```python
import gradio as gr
from gtts import gTTS
from transformers import pipeline
import tempfile
import subprocess

# Load pipelines once (faster inference)
asr = pipeline("automatic-speech-recognition", model="openai/whisper-small")
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-fr")

def translate_speech(text_input):
    # Step 1: Text to English speech (MP3)
    tts_en = gTTS(text=text_input, lang='en')
    en_mp3 = tempfile.NamedTemporaryFile(delete=False, suffix=".mp3")
    tts_en.save(en_mp3.name)

    # Step 2: Convert to WAV (required for Whisper ASR)
    en_wav = tempfile.NamedTemporaryFile(delete=False, suffix=".wav")
    subprocess.call([
        "ffmpeg", "-i", en_mp3.name, "-ar", "16000", "-ac", "1", "-y", en_wav.name
    ], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

    # Step 3: Transcribe speech
    result = asr(en_wav.name)
    recognized_text = result['text']

    # Step 4: Translate to French
    translated = translator(recognized_text)[0]['translation_text']

    # Step 5: French speech synthesis
    tts_fr = gTTS(text=translated, lang='fr')
    fr_mp3 = tempfile.NamedTemporaryFile(delete=False, suffix=".mp3")
    tts_fr.save(fr_mp3.name)

    return recognized_text, translated, en_mp3.name, fr_mp3.name

# Gradio UI
iface = gr.Interface(
    fn=translate_speech,
    inputs=gr.Textbox(label="Enter English Text", placeholder="Type something..."),
    outputs=[
        gr.Textbox(label="Recognized English Speech"),
        gr.Textbox(label="Translated French Text"),
        gr.Audio(label="English Speech"),
        gr.Audio(label="French Speech")
    ],
    title="🗣️ English-to-French Speech Translator",
    description="Converts English text to speech, transcribes it, translates to French, and plays both audios."
)

iface.launch()
```
